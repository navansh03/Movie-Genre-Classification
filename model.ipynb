```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import pickle
```

```python
# Step 2: Load Dataset
df_train=pd.read_csv("train_data.txt",sep=":::",header=0,engine="python")
```

```python
df_train.head()
```

```python
df_train.columns=['SN','movie_name','genre','confession']
```

```python
df_train.head()
```

```python
df_train['cleaned_plot'] = df_train['confession'].str.lower()
```

```python
df_train.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train['cleaned_plot'], df_train['genre'], test_size=0.2, random_state=42)
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

## Naive Bayes Model Training


```python

naive_bayes_model=MultinomialNB() 
naive_bayes_model.fit(X_train_vec, y_train)
```

## Logistic Regression Model

```python
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(X_train_vec, y_train) 
```

## Model Eval

```python
y_pred1 = naive_bayes_model.predict(X_test_vec)
y_pred2 = logistic_regression_model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, y_pred1))
print(classification_report(y_test, y_pred1))
print("Accuracy of Logistic Regression:", accuracy_score(y_test, y_pred2))
print(classification_report(y_test, y_pred2))
```

#### Save Model and Vectorizer to PKL file to use in the app

```python
with open('movie_genre_classifier_NB.pkl', 'wb') as file:
    pickle.dump(naive_bayes_model, file)
with open('movie_genre_classifier_LR.pkl', 'wb') as file:
    pickle.dump(logistic_regression_model, file)
with open('tfidf_vectorizer.pkl', 'wb') as file:
    pickle.dump(vectorizer, file)
```
